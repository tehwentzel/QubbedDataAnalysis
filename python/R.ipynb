```r
library(stats)
library(tidyverse)
library(dplyr)
data <- read.csv('../data/dose_cluster_03032022.csv')
head(data)
```

```r
symptom_endpoint_cols <- list()
dose_cluster_cols <- list()
for(name in names(data)){
    if(grepl('_max_',name,fixed=TRUE) | grepl('digest',name,fixed=TRUE)){
        symptom_endpoint_cols[length(symptom_endpoint_cols)+1] <- name
    }
    else if(grepl('_dose_cluster',name,fixed=TRUE)){
        dose_cluster_cols[length(dose_cluster_cols)+1] <- name
    }
}
pvals <- c()
x <- c()
y <- c()
for(dcol in dose_cluster_cols){
    for(scol in symptom_endpoint_cols){
        ftest <- fisher.test(table(data[[dcol]],data[[scol]]),simulate.p.value=TRUE,workspace=2e8)
        ftest <- ftest[['p.value']]
        if(ftest < .5){
            pvals[length(pvals)+1] <- ftest
            x[length(x)+1] <- dcol
            y[length(y)+1] <- scol
        }
    }
}
correlations <- data.frame(pvals,x,y)
correlations <- correlations[order(pvals),]
correlations[which(correlations$pvals < .05),]
```
